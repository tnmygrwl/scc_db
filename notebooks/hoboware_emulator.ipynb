```python
# Import dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the data from the csv file

CONDUCTIVITY_DATA = '../data/RB1_ConDL_6_9_2023_AH.csv'
con_data = pd.read_csv(CONDUCTIVITY_DATA)
con_data.head()
```

```python
# Filter the data to get the desired rows and columns
# Note: negative number in the end index means that the index is counted from the end of the data
# e.g. -1 means the last row or column
row_start = 8
row_end = -2
col_start = 1
col_end = -2


def filter_data(con_data, row_start, row_end, col_start, col_end):
    # Adjust for 1-indexing if end indices are positive
    row_start -= 1 if row_start > 0 else 0
    col_start -= 1 if col_start > 0 else 0

    # Get the data
    con_data = con_data.iloc[row_start:row_end, col_start:col_end]
    return con_data

con_data = filter_data(con_data, row_start, row_end, col_start, col_end)
con_data.head()


# Note: Re-running this cell will overwrite the data in the csv file and keep filtering the data, 
# a good practice is to only run it after running the previous cell and reading the original file again
```
